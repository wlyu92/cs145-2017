```python
plt.plot(df['total_rooms'] / df['households'], df['median_income'], 'o', alpha=.25)
plt.xlabel("rooms per household")
plt.ylabel("median income")
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
regr = linear_model.LinearRegression()
regr.fit((df["total_rooms"] / df['households']).reshape(-1, 1), df["median_income"].reshape(-1, 1))
print regr.coef_
print regr.score((df["total_rooms"] / df['households']).reshape(-1, 1), df["median_income"].reshape(-1, 1))
```

```python
plt.plot(df['total_rooms'] / df['households'], df['median_income'], 'o', alpha=.25)
plt.xlabel("rooms per household")
plt.ylabel("median income")
xmin, xmax = plt.gca().get_xlim()
ymin, ymax = plt.gca().get_ylim()
plt.plot([xmin, xmax], [regr.predict(xmin)[0], regr.predict(xmax)[0]])
plt.gca().set_ylim(ymax=ymax)
```
